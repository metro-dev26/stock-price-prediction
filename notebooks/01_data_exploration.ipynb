```python
# Cell 1: Import libraries
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully!")
```

```python
# Cell 2: Download data
# Let's use Reliance Industries (Indian stock)
ticker = 'RELIANCE.NS'

# Download 5 years of data
print(f"Downloading data for {ticker}...")
data = yf.download(ticker, start='2019-02-01', end='2024-02-01')

print(f"Data downloaded! Shape: {data.shape}")
data.head()
```

```python
# Cell 3: Save data
data.to_csv('../data/stock_data.csv')
print("Data saved to data/stock_data.csv")
```

```python
print("Dataset Info:")
print(data.info())
print("\nBasic Statistics:")
print(data.describe())
print("\nMissing values:")
print(data.isna().sum())
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], linewidth=2)
plt.title(f'{ticker} Stock Price (2019-2024)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price ()', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../results/price_history.png', dpi=300, bbox_inches='tight')
plt.show()

print("Chart saved to results/price_history.png")
```
