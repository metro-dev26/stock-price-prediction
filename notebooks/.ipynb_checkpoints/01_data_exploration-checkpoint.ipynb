```python
# Cell 1: Import libraries
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully!")
```

```python
# Cell 2: Download data
# Let's use Reliance Industries (Indian stock)
ticker = 'RELIANCE.NS'

# Download 5 years of data
print(f"Downloading data for {ticker}...")
data = yf.download(ticker, start='2019-02-01', end='2024-02-01')

print(f"Data downloaded! Shape: {data.shape}")
data.head()
```

```python
# Cell 3: Save data
data.to_csv('../data/stock_data.csv')
print("Data saved to data/stock_data.csv")
```

```python
print("Dataset Info:")
print(data.info())
print("\nBasic Statistics:")
print(data.describe())
print("\nMissing values:")
print(data.isna().sum())
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], linewidth=2)
plt.title(f'{ticker} Stock Price (2019-2024)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price ()', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../results/price_history.png', dpi=300, bbox_inches='tight')
plt.show()

print("Chart saved to results/price_history.png")
```

```python
# Cell 6: Calculate daily returns
data['Returns'] = data['Close'].pct_change()

# Plot returns
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

ax1.plot(data.index, data['Close'])
ax1.set_title('Stock Price', fontsize=14)
ax1.set_ylabel('Price ()')
ax1.grid(True, alpha=0.3)

ax2.plot(data.index, data['Returns'], color='orange', alpha=0.7)
ax2.set_title('Daily Returns', fontsize=14)
ax2.set_xlabel('Date')
ax2.set_ylabel('Returns')
ax2.grid(True, alpha=0.3)
ax2.axhline(y=0, color='r', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig('../results/price_and_returns.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Cell 7: Returns distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data['Returns'].dropna(), bins=50, alpha=0.7, edgecolor='black')
plt.title('Distribution of Daily Returns', fontsize=14)
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.boxplot(data['Returns'].dropna())
plt.title('Returns Boxplot', fontsize=14)
plt.ylabel('Returns')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../results/returns_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Cell 8: Volume analysis
plt.figure(figsize=(14, 5))
plt.plot(data['Volume'])
plt.title('Trading Volume Over Time', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Volume')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../results/volume.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Cell 9: Summary statistics
print("="*60)
print("STOCK ANALYSIS SUMMARY")
print("="*60)
print(f"\nTicker: RELIANCE.NS")
print(f"Total trading days: {len(data)}")

print("\n" + "="*60)
print("PRICE STATISTICS")
print("="*60)
print(data['Close'].describe())

print("\n" + "="*60)
print("RETURNS STATISTICS")
print("="*60)
print(data['Returns'].describe())

print("\n" + "="*60)
print("VOLUME STATISTICS")
print("="*60)
print(data['Volume'].describe())
print("="*60)
```
